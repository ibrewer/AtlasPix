```python
# SORT
data = open("TOT_test_5.txt", 'r')
total_lines = data.readlines()
hist_data = np.empty((len(total_lines), 23))
counter = 0
for line in total_lines:
    entry = np.array(line.strip().split(), dtype=float)
    if len(entry[2:len(entry)]) == 21:
        x = int(entry[0])
        y = int(entry[1])
        hist_data[counter] = np.append([x,y], entry[2:len(entry)])
        counter += 1
    
print(hist_data)
```
